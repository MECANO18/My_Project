```python
"""1. Écrire un programme qui permet de transformer le fichier examens.xlsx en DataFrame."""

P = pd.ExcelFile("examens.xlsx")
P = P.parse('Resultats')
P
```

```python
"""2. En se basant sur les tableaux des coefficients, écrire un programme qui cacule la moyenne
de chaque UE et détermine si l’étudiant l’a validée ou pas."""
#Df des coefs
Coefs=[2.5,2.5,1,2.5,2.5,2.5,2.5,1,2.5,2.5,1.5,1.5,1,2,2]
EC = ['Phy1', 'Phy2', 'TPPhy', 'Math1', 'Math2'
      , 'Chimie1' , 'Chimie2' , 'TPChimie', 'Info1'
      , 'Info2' , 'Base1' , 'Base2' , 'Base3' , 'Gr1' , 'Gr2' ]
C = pd.DataFrame(Coefs , EC)[0]

#UE_PHYSIQUE
ue_phy = P.iloc[:,2:5]
ue_phy.Physique1 = ue_phy.Physique1 * C['Phy1']
ue_phy.Physique2 = ue_phy.Physique2 * C['Phy2']
ue_phy.TP_Physique = ue_phy.TP_Physique * C['TPPhy']
Moy_phy= np.sum(ue_phy, axis = 1)/(C['Phy1']+C['Phy2']+C['TPPhy'])
Moy_phy
crdt_phy = []
dec_phy = []
for i in range (len(Moy_phy)):
    if Moy_phy[i] >=10:
        crdt_phy.append(12)
        dec_phy.append('V')
    else:
        crdt_phy.append(0)
        dec_phy.append('NV')

#UE_MATHS
ue_maths = P.iloc[:,5:7]
ue_maths.Maths1 = ue_maths.Maths1 * C['Math1']
ue_maths.Maths2 = ue_maths.Maths2 * C['Math2']
Moy_maths= np.sum(ue_maths, axis = 1)/(C['Math1']+C['Math2'])
crdt_maths = []
dec_maths = []
for i in range (len(Moy_maths)):
    if Moy_maths[i] >= 10 :
        crdt_maths.append(10)
        dec_maths.append('V')
    else:
        crdt_maths.append(0)
        dec_maths.append('NV')
        
#UE_CHIMIE
ue_chimie = P.iloc[:,7:10]
ue_chimie.Chimie1 = ue_chimie.Chimie1 * C['Chimie1']
ue_chimie.Chimie2 = ue_chimie.Chimie2 * C['Chimie2']
ue_chimie.TP_Chimie = P.TP_Chimie.values * C['TPChimie']
Moy_chimie = np.sum(ue_chimie, axis = 1)/(C['Chimie1']+C['Chimie2']+C['TPChimie'])
crdt_chimie = []
dec_chimie = []
for i in range (len(Moy_chimie)):
    if Moy_chimie[i] >= 10 :
        crdt_chimie.append(12)
        dec_chimie.append('V')
    else:
        crdt_chimie.append(0)
        dec_chimie.append('NV')        

#UE_INFO
ue_info = P.iloc[:,10:12]
ue_info.Info1=ue_info.Info1 * C['Info1']
ue_info.Info2=ue_info.Info2 * C['Info2']
Moy_info= np.sum(ue_info, axis = 1)/(C['Info1']+C['Info2'])
crdt_info = []
dec_info = []
for i in range (len(Moy_info)):
    if Moy_info[i] >= 10 :
        crdt_info.append(10)
        dec_info.append('V')
    else:
        crdt_info.append(0)
        dec_info.append('NV')

#UE_BASE
ue_base = P.iloc[:,12:15]
ue_base.Base1=ue_base.Base1 * C['Base1']
ue_base.Base2=ue_base.Base2 * C['Base2']
ue_base.Base3=ue_base.Base3 * C['Base3']
Moy_base = np.sum(ue_base, axis = 1)/(C['Base1']+C['Base2']+C['Base3'])
crdt_base = []
dec_base = []
for i in range (len(Moy_base)):
    if Moy_base[i] >= 10 :
        crdt_base.append(8)
        dec_base.append('V')
    else:
        crdt_base.append(0)
        dec_base.append('NV')

#UE_GRAPHIQUE
ue_gr = P.iloc[:,15:]
ue_gr.Graphique1=ue_gr.Graphique1 * C['Gr1']
ue_gr.Graphique2=ue_gr.Graphique2 * C['Gr2']
Moy_gr= np.sum(ue_gr, axis = 1)/(C['Gr1']+C['Gr2'])
crdt_gr = []
dec_gr = []
for i in range (len(Moy_gr)):
    if Moy_gr[i] >= 10 :
        crdt_gr.append(8)
        dec_gr.append('V')
    else:
        crdt_gr.append(0)
        dec_gr.append('NV')
```

```python
"""Écrire un programme qui insère une colonne de validation (V : UE validée, NV : UE non validée) après chaque UE."""
P.insert(5, "Stat_phy", dec_phy, allow_duplicates = False)
P.insert(8, "Stat_maths", dec_maths, allow_duplicates = False)
P.insert(12, "Stat_chimie", dec_chimie, allow_duplicates = False)
P.insert(15, "Stat_info", dec_info, allow_duplicates = False)
P.insert(19, "Stat_base", dec_base, allow_duplicates = False)
P.insert(22, "Stat_gr", dec_gr, allow_duplicates = False)
```

```python
"""Écrire un programme qui trie les noms et prénoms par ordre alphabétique."""
P = P.sort_values(by = ["Nom", "Prénom"], ascending = [True, True])
P
```

```python
"""Écrire un programme qui enregistre dans un nouveau fichier excel les résultats obtenus."""
R = pd.ExcelWriter("Result_examens.xlsx", engine = "xlsxwriter")
P.to_excel(R, "Déliberation", index = False)
R.save()
```
